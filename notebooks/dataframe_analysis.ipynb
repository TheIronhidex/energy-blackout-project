```python
list_1 = df1['country_name']
list_2 = df8['country']
exclusion_list = np.setdiff1d(list_2,list_1)
```

```python
df8_mod = df8[~df8['country'].isin(exclusion_list)]
```

```python
df1.rename(columns={"country_name": "country"}, inplace=True)
```

```python
df1.set_index('country', inplace=True)
```

```python
df_last = df1.merge(df8_mod, how='inner', on='country')
```

```python
df_last.set_index('year',drop=True,inplace=True)
df_last = df_last.groupby(['country'],as_index = True).mean()
```

```python
df_last.isna().sum()
```

```python
df_last.dropna(axis=1,how='any',inplace=True)
```

```python
df_last.to_csv('data/energy-var.csv' , index = True)
```
